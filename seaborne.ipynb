```python
from sklearn.datasets import fetch_california_housing
```

```python
house_california=fetch_california_housing()
```

```python
house=pd.DataFrame(house_california.data,columns=house_california.feature_names)
```

```python
house.head()
```

```python
house['Price']=house_california.target
```

```python
house.head()
```

```python
print(house_california)
```

```python
#Distribution Plot
sb.histplot(house['Price'])
```

```python
sb.distplot(house['Price'])
```

```python
#Correlation
#Positive
#Negative
```

```python
correlation=house.corr()
```

```python
#Construct heat map
fig=plt.figure(figsize=(10,10))
sb.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')

```
